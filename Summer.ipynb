```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
```

# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
df["City"].unique()
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []
gs = df[df['Medal'] == 'Gold']
for Sport in gs['Sport'].unique():
    data.append([Sport , len(gs[gs['Sport']  == Sport])])
    
pd.DataFrame(data,columns = ['Sports','Golds']).sort_values(by='Golds', ascending=False).head().plot(x = 'Sports', y = 'Golds', kind = 'bar', figsize = (6,5), color = "red")
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
most = []

for Sport in df['Sport'].unique():
    most.append([Sport , len(df[df['Sport']  == Sport])])
pd.DataFrame(most,columns = ['Sports','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sports', y = 'Medals', kind = 'bar', figsize = (6,5), color = "green")
```

### 4. Which player has won most number of medals? (Top 5)

```python
most = []
for Athlete in df['Athlete'].unique():
    most.append([Athlete , len(df[df['Athlete']  == Athlete])])

pd.DataFrame(most,columns = ['Athlete','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Athlete', y = 'Medals', kind = 'bar', figsize = (6,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
most = []
for Event in df['Event'].unique():
    most.append([Event , len(df[df['Event']  == Event])])

pd.DataFrame(most,columns = ['Event','Players']).sort_values(by='Players', ascending=False).head().plot(x = 'Event', y = 'Players', kind = 'bar', figsize = (6,5),color='purple')
```
